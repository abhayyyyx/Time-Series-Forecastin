# Checking if the data is stationary or not

To check that, we can adfuller library

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series):
    result = adfuller(series,autolag='AIC')
    labels = ['ADF Test Statistic','p-value','#Lags used','Number of Observation Used']
    for value, label in zip(result,labels):
        print(f'{label}: {value}')
    if result[1] < 0.05:
        print("Strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("Weak evidence aganist the null hypothesis, fail to reject the null hypothesis.Data has a unit root and is non-stationary")
```

```python
#Performing the ADF test on average price
print("ADF test on average price:")
adf_test(data['Average Price'].dropna())
```

```python
def adf_test(series):
    result = adfuller(series.dropna(),autolag='AIC')
    labels = ['ADF test statistic','p-value','#lag used','Number of observation used']
    for value, label in zip(result,labels):
        print(f'{label} : {value}')
    if result[1] < 0.05:
        print('Strong evidence against the null hypotheis,reject the null hypothesis,data is stationary')
    else:
        print('Weak evidence against the null hypothesis, failed to reject the null hypothesis, Data is not stationary')

print("ADF test on original data")
adf_test(data['Average Price'])
```

```python
#Function to plot rolling statistics

def plot_rolling_statistic(series, title = "Rolling mean and standard deviation"):
    plt.figure(figsize=(12,8))
    plt.plot(series,color='blue',label = 'Original')
    plt.plot(series.rolling(window=12).mean(),color='red',label='Rolling Mean')
    plt.plot(series.rolling(window=12).std(),color='black',label='Rolling Standard Deviation')
    plt.legend(loc='best')
    plt.title(title)
    plt.show()
```

```python
plot_rolling_statistic(data['Average Price'],title='Rolling mean & standard deviation of average prices')
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

#plotting ACF and PACF 

plt.figure(figsize=(12,8))
plot_acf(data['Average Price'],lags=50)
plt.title("Autocorrelation function (ACF)")
plt.show()

plt.figure(figsize=(12,8))
plot_pacf(data['Average Price'],lags=50)
plt.title("Partial Autocorrelation Function (PACF)")
plt.show()
```

# Performing Kwiatkowski-Phillips-Schmidt-Shin KPSS - Test

```python
from statsmodels.tsa.stattools import adfuller, kpss

def kpss_test(series, **kw):
    statistic, p_value, n_lags, crictical_values = kpss(series,**kw)
    print(f'KPSS statistic: {statistic}')
    print(f'P value : {p_value}')
    print(f'Number of lags : {n_lags}')
    print('Critical values:')
    for key, value in crictical_values.items():
        print(f' {key} : {value}')
    if p_value < 0.05:
        print("Strong evidence against null hypothesis,reject the null hypothesis. Data is Stationary")
    else:
        print("Weak evidence against null hypothesis, failed to reject the null hypothesis. Data is non-stationary")

print("KPSS test:")
kpss_test(data['Average Price'].dropna(),regression='c')
```

# Second attempt to de-tredning the data

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Differencing the data, dropping the first NaN value
data['Differencing'] = data['Average Price'].diff().dropna()

# Exponential Smoothing 
exp_smooth = ExponentialSmoothing(data['Average Price'].dropna(), trend='additive', seasonal=None).fit()
data['Exp Smooth'] = exp_smooth.fittedvalues

# Dropping NaN values for all columns
data_cleaned = data.dropna(subset=['Average Price', 'Differencing', 'Exp Smooth'])

# Performing ADF, KPSS on original, differenced, and smoothed data
print("ADF test on Original Average Price:")
adf_test(data_cleaned['Average Price'])
print("\nKPSS test on Original Average Price:")
kpss_test(data_cleaned['Average Price'], regression='c')

print("\nADF test on Differenced Price:")
adf_test(data_cleaned['Differencing'])
print("\nKPSS test on Differenced Price:")
kpss_test(data_cleaned['Differencing'], regression='c')

print("\nADF test on Exponentially Smooth Price:")
adf_test(data_cleaned['Exp Smooth'])
print("\nKPSS test on Exponentially Smooth Price:")
kpss_test(data_cleaned['Exp Smooth'], regression='c')
```

```python
#plot the original, differened, exponential smooth data 
plt.figure(figsize=(10,8))
plt.plot(data_cleaned['Average Price'],label = 'Original price',color='black')
plt.plot(data_cleaned['Differencing'],label='Differenced price')
plt.plot(data_cleaned['Exp Smooth'],label='Exponential Smooth price')
plt.legend(loc='best')
plt.title("Average price of original,differenced,exponential smooth data")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Define the stock symbol and the date range
symbol = 'RELIANCE.NS'
start_date = '2014-07-01'
end_date = '2024-06-30'

# Fetch the stock data
data = yf.download(symbol, start=start_date, end=end_date)

# Calculate the average price (Open, Low, High, Adj Close)
data['Avg Price'] = data[['Open', 'Low', 'High', 'Adj Close']].mean(axis=1)

# Differencing to remove trend
data['Differenced'] = data['Avg Price'].diff()
data.dropna(subset=['Differenced'], inplace=True)  # Drop NA values created by differencing

# Split data into training and testing sets
train_size = int(len(data) * 0.8)
train = data['Avg Price'][:train_size]
test = data['Avg Price'][train_size:]

# Fit ARIMA model
arima_model = ARIMA(train, order=(1, 1, 1))  # Order (p, d, q) needs to be tuned
arima_fitted_model = arima_model.fit()

# Make predictions with ARIMA
arima_predictions = arima_fitted_model.forecast(steps=len(test))

# Fit SARIMAX model
sarimax_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))  # Seasonal order (p, d, q, s) needs to be tuned
sarimax_fitted_model = sarimax_model.fit()

# Make predictions with SARIMAX
sarimax_predictions = sarimax_fitted_model.forecast(steps=len(test))

# Fit Holt-Winters model
holt_winters_model = ExponentialSmoothing(train, trend='add', seasonal='add', seasonal_periods=12)
holt_winters_fitted_model = holt_winters_model.fit()

# Make predictions with Holt-Winters
holt_winters_predictions = holt_winters_fitted_model.forecast(steps=len(test))

# Calculate evaluation metrics
arima_mae = mean_absolute_error(test, arima_predictions)
arima_mse = mean_squared_error(test, arima_predictions)
arima_rmse = np.sqrt(arima_mse)

sarimax_mae = mean_absolute_error(test, sarimax_predictions)
sarimax_mse = mean_squared_error(test, sarimax_predictions)
sarimax_rmse = np.sqrt(sarimax_mse)

holt_winters_mae = mean_absolute_error(test, holt_winters_predictions)
holt_winters_mse = mean_squared_error(test, holt_winters_predictions)
holt_winters_rmse = np.sqrt(holt_winters_mse)

print(f"ARIMA Model - MAE: {arima_mae}, MSE: {arima_mse}, RMSE: {arima_rmse}")
print(f"SARIMAX Model - MAE: {sarimax_mae}, MSE: {sarimax_mse}, RMSE: {sarimax_rmse}")
print(f"Holt-Winters Model - MAE: {holt_winters_mae}, MSE: {holt_winters_mse}, RMSE: {holt_winters_rmse}")

# Future prediction
future_steps = 30
future_dates = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=future_steps, freq='B')  # 'B' for business days

# Forecast future values with ARIMA, SARIMAX, and Holt-Winters models
arima_future = arima_fitted_model.forecast(steps=future_steps)
sarimax_future = sarimax_fitted_model.forecast(steps=future_steps)
holt_winters_future = holt_winters_fitted_model.forecast(steps=future_steps)

# Plot the results
plt.figure(figsize=(14, 7))

# Historical data
plt.plot(data.index, data['Avg Price'], label='Historical Data')

# Test data
plt.plot(test.index, test, label='Test Data', color='blue')

# ARIMA Predictions
plt.plot(test.index, arima_predictions, label='ARIMA Predictions', color='red')

# SARIMAX Predictions
plt.plot(test.index, sarimax_predictions, label='SARIMAX Predictions', color='green')

# Holt-Winters Predictions
plt.plot(test.index, holt_winters_predictions, label='Holt-Winters Predictions', color='purple')
```

```python
#Future Predictions
plt.plot(future_dates,arima_future,label = 'ARIMA Future Prediction',color='orange',linestyle='--')
plt.plot(future_dates,sarimax_future,label = 'SARIMAX Future Predictions',color='blue',linestyle = '--')
plt.plot(future_dates,holt_winters_future,label ='HOLT-WINTERS Predictions',color='purple',linestyle='--')

plt.legend(loc='best')
plt.title("ARIMA, SARIMAX and HOLT-WINTERS Model Predictions")
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
